### 1. What is name of car that has the highest horsepower?  

```python
max_hp_row = df.loc[df['horsepower'].idxmax()]
print("Car with highest horsepower:", max_hp_row['car_name'])
print("Horsepower:", max_hp_row['horsepower'])
```

### 2. How many cars have mpg ≥ 35?

```python
num_cars_high_mpg = (df['mpg'] >= 35).sum()

print("Number of cars with mpg ≥ 35:", num_cars_high_mpg)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered_df = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

most_common_origin = filtered_df['origin'].mode().iloc[0]

print("Most common origin for cars with horsepower > 100 and weight < 3000:", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
df['origin'] = df['origin'].str.strip().str.lower()
mean_acc_japan = round(df[df['origin'] == 'japan']['acceleration'].mean(), 2)
print("Mean acceleration of cars from Japan :", mean_acc_japan)
```

### 5. Which year had the highest average mpg?

```python
year_highest_avg_mpg = df.groupby('model_year')['mpg'].mean().idxmax()
print("Year with the highest average mpg : ", year_highest_avg_mpg)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
df.reset_index(inplace=True)
df.set_index('car_name', inplace=True)

median_mpg = df['mpg'].median()
above_median = df[df['mpg'] > median_mpg]
best_cars = above_median[above_median['hp_to_weight'] == above_median['hp_to_weight'].max()]
best_hp_to_weight_car_names = best_cars.index.tolist()
print("Cars with the best ratio of horsepower to weight among all cars that also have above-median mpg : ", best_hp_to_weight_car_names)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
# Group by year and origin to get average mpg
avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plot using Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg_by_year_origin, x='model_year', y='mpg', hue='origin', marker='o')

# Customizing the plot
plt.title('Average MPG Over the Years by Origin', fontsize=16)
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()

plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
df.reset_index(drop=True, inplace=True)

plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(20, 200), # sizes=(20, 200) ensures size by mpg is visually meaningful.
	alpha=0.7,       # alpha=0.7 adds transparency for overlapping points.
	edgecolor='gray' # edgecolor='gray' gives each point definition.

)

# Titles and labels
plt.title('Horsepower vs. Weight of Cars (Sized by MPG, Colored by Origin)', fontsize=14)
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)

plt.show()
```
